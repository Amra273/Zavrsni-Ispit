# **4. Zadatak - Machine Learning**


Za dostavljeni Iris dataset odaberite klasifikacijski model kojim ce biti moguce sto bolje na osnovu features u datasetu predvidjeti klasu biljke. 


Zadatak ce biti ocijenjen po sljedecim kriterijima:


*   Odabir machine learning modela (dodati bodovi za komentare, analizu, vizualizaciju i pregledavanje podataka i elaboraciju zasto je dati model odabran)
*   Kod (urednost, logicnost, modularnost)

*   Odabir features, testiranje, metrike (dodatni bodovi za vizualizacije)


```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns



#ucitavamo dataset u pandas 
data = pd.read_csv("Iris.csv")

#problem predstavljaju nan vrijednosti, pa ih ovom linijom koda mijenjamo sa nulom
data = data.fillna(0)
```

```python
#Biramo feature-e i labele
X = data.iloc[:, 1:5].values
y = data.iloc[:, 5].values

from sklearn.preprocessing import LabelEncoder
le = LabelEncoder()
y = le.fit_transform(y)
```

```python
#normalizacija podataka
from sklearn.model_selection import train_test_split
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size = 0.2, random_state = 0)
```

```python
# biranje klasifikatora
# ucitavanje potrebnih biblioteka
from sklearn.neighbors import KNeighborsClassifier
from sklearn.metrics import confusion_matrix, accuracy_score
from sklearn.model_selection import cross_val_score

# Kreiranje modela, pocinjemo sa uobicajenim n_neighbors = 5 (default)
classifier = KNeighborsClassifier(n_neighbors=5)

# Fitovanje modela
classifier.fit(X_train, y_train)

# Predviđanje rezultata
y_pred = classifier.predict(X_test)

```

```python
#kreiranje matrice koja upoređuje stvarne ciljne vrijednosti sa vrijednostima koje su pretpostavljenje pomoću ML-a
cm = confusion_matrix(y_test, y_pred)
cm

#Tačnost modela
accuracy = accuracy_score(y_test, y_pred)*100
print('Tačnost modela iznosi ' + str(round(accuracy, 2)) + ' %.')

# lista koraka za KNN
k_list = list(range(1,50,2))
# cross validation 
cv_scores = []

#procjena prediktivnih modela
for k in k_list:
    knn = KNeighborsClassifier(n_neighbors=k)
    scores = cross_val_score(knn, X_train, y_train, cv=10, scoring='accuracy')
    cv_scores.append(scores.mean())

#greska
MSE = [1 - x for x in cv_scores]
```

```python
#VIZUALIZACIJA

plt.figure()
plt.figure(figsize=(15,10))
plt.title('Optimalan broj susjeda', fontsize=20, fontweight='bold')
plt.xlabel('Broj susjeda k', fontsize=15)
plt.ylabel('Greška klasifikacije', fontsize=15)
sns.set_style("whitegrid")
plt.plot(k_list, MSE)

plt.show()
```

```python
# nalaženje najboljeg broja k
best_k = k_list[MSE.index(min(MSE))]
print("Optimalan broj susjeda(neighbor-a) je %d." % best_k)
```
